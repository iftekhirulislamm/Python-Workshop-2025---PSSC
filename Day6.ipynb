# Dictionary

Dictionary store values as (Key:Value) Pair. Values can be accessed through keys. Values can be changed but not keys.

values can be of every data type. But list can not be keys since list is mutable but not keys


```python
student_details = {
    "dept":"EEE",
    "id": 34,
    "name": "Sakib alam",
    "address": ["walton mess","balur math"],
    "Favourite_Subject": ("Electromagnetics","Circuit"),
    ("Fav_Topic","Fav_Book"):"Bola jabena"
}

print(student_details)
```

```python
print(type(student_details))
```

# Values can be accessed with keys

```python
print(student_details["id"])
print(student_details[("Fav_Topic","Fav_Book")])
```

Changing values for dictionary

```python
student_details["id"]=534
print(student_details)
```

# Nested Dictionary

```python
student_details = {
    "dept": "EEE",
    "id": 34,
    "name": "Sakib alam",
    "address": ["walton mess", "balur math"],
    "Favourite_Subject": ("Electromagnetics", "Circuit"),
    ("Fav_Topic", "Fav_Book"): "Bola jabena",
    "Favourite Club":  { "Name":"PSSC",
                               "Location": "GSTU"}
}

print(student_details)
```

```python
print(student_details["Favourite Club"]["Name"])
```

```python
print(type(student_details["Favourite Club"]))
print(type(student_details["Favourite Club"]["Location"]))
```

# Set

Collection of unordered items. Each element is unique and immutable. List an Dictionaries can not be stored since they are mutable

```python
a = {1,3,2,4}
b = {1,2,8}

print(a)
print(b)
```

# While taking input set ignores duplicate values

```python
x={1,2,3,3,4,5,6,6}
print(x)
```

# Different operations on Set

```python
a = {1,3,2,4}
b = {1,2,8}

print(a)
print(b)
```

```python
print(a-b)
print(a&b) # Intersection
print(a|b) # Union
```
