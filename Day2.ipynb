Reversing a 4 digit number

```python
x=int(input("Enter a number: "))
result=x%10             #result =4, x=1234
x=x//10                 #result =4, x=123
result=result*10+x%10   #result =43, x= 123
x=x//10                 #result =43, x=12
result=result*10+x%10   #result =432, x=12
x=x//10                 #result =432, x=1
result=result*10+x%10   #result =4321, x=1
print(result)
```
