# Grade Calculator

```python
#1st problem
number =int(input())

if(number>=80 and number<=100):
  print("A+")
elif(number>=70 and number<80):
  print("A")
elif(number>=60 and number<70):
  print("B")
elif(number>=50 and number<60):
  print("C")
elif(number>=40 and number<50):
  print("D")
else:
    print("F")
```

# Finding largest among 3 numbers

```python
#3rd problem
a=int(input())
b=int(input())
c=int(input())

if(a>b):
  if(a>c):
    print(a)
  else:
    print(c)
else:
  if(b>c):
    print(b)
  else:
    print(c)
```

```python
#4th problem
a=int(input())
b=int(input())
c=input()

if(c=='+'):
    print(a+b)
elif(c=='-'):
    print(a-b)
elif(c=='*'):
    print(a*b)
elif (c=='/'):
    print(a/b)
else:
    print("Incorrect input")

```

# in keyword

```python
x="pssn foef"
if ('  ') in x:
  print("present")
else:
  print("not present")
```

```python
x="pssn  foef"
if ('  ') in x:
  print("present")
else:
  print("not present")
```

# Checking first and last character

```python
x="pssc"

if(x[0]==x[-1]):
  print("same")
else:
  print("not same")
```

# List

Accepts different data types

```python
list1=[1,2,3,4,'sakib']
print(list1)
```

```python
list1=[1,2,3,4,'sakib',(1,2,3,4,'sakib')]
print(list1)
```

```python
print(type(list1))
```

Indexing

```python
print(list1[1])
```

Nested List

```python
list2=[1,2,4,5,[1,2,[1,2,3],34]]
print(list2[4])
```

```python
print(list2[4][2])
```

```python
print(list2[4][2][2])
```

Values can be changed

```python
list2[4][2][2]=[9,9]
print(list2)
```

Slicing can be done as strings

```python
print(list2[:2])
```

```python
list3=[1,2,(3,4)]
print(list3)
```

```python
list3[2]=[1,2,(3,4)]
print(list3)
```
