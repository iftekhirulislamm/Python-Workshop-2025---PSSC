# Summation of the digits of a two digit number

```python
x=int(input("Enter a two digit number:"))
sum=0
second_digit=x%10
sum+=second_digit
x=x//10
first_digit=x%10
sum+=x
print(f"{first_digit}+{second_digit}={sum}")
```

# Positive index & Negative index
positive index starts with 0 from the left character.
Negative index starts with -1 from the most right character

```python
x='Programming Club'
print(x[1])
print(x[-1])
print(x[5])
print(x[-2])
print(x[-15])
```

# Slicing
slicing means cutting a portion of the string. **variable_name[Start_index : End_index]** This will extract from start index to just before ending index

```python
x='Programming Club'
print(x[0:5])
```

```python
x='Programming Club'
print(x[2:16])
```

No start index means 0. no ending index means till last

```python
x='Programming Club'
print(x[:16])
```

```python
x='Programming Club'
print(x[0:])
```

```python
x='Programming Club'
print(x[:])
```

```python
x='Programming Club'
print(x[5:])
```

# Conditional Statement

if else

```python
x=10
if(x>9):
    print("x is Greater")
else:
    print("x is not Greater")
```

if elif else

```python
x=10
if(x>10):
    print("x is Greater than 10")
elif(x<10):
    print("x is not Greater than 10")
else:
    print("x is equal to 10")
```

Nested if else

```python
x=13
if(x>10):
    if(x%2==0):
        print("x is even and larger than 10")
    else:
        print("x is not even but larger than 10")
elif(x<10):
    print("x is not Greater than 10")
else:
    print("x is equal to 10")
```
